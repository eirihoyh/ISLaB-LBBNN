```python
import numpy as np

median_res_abalone_lrt = np.loadtxt("results/flow_class_skip_2_hidden_200_dim_5000_epochs_0.01_lr_abalone_sigmoid_median.txt", delimiter=",")
full_res_abalone_lrt = np.loadtxt("results/flow_class_skip_2_hidden_200_dim_5000_epochs_0.01_lr_abalone_sigmoid_full.txt", delimiter=",")
```

```python
import numpy as np
import matplotlib.pyplot as plt
alpha_levels = np.array(["0.001-0.01","0.01-0.1","0.1-0.25","0.25-0.5","0.5-0.75","0.75-0.9","0.9-0.99","0.99-0.999"])

median_res_abalone_lrt[:,-1] = median_res_abalone_lrt[:,-1]**0.5

names = ["RMSE", "Nr. of used weights in MPM", "Correlation"]

for i in range(3):
    fig, ax = plt.subplots()
    x_values = np.arange(len(alpha_levels))
    ax.plot(x_values, median_res_abalone_lrt[:,i],"-o")
    
    # Set the x-tick locations (optional, if you want specific positions)
    ax.set_xticks(x_values) 
    ax.set_ylabel(names[i])
    ax.set_xlabel("Inital alpha value range")
    # Set the new labels for the x-ticks
    ax.set_xticklabels(alpha_levels, rotation=60)
    plt.grid()
    plt.title("Median probability model")
    plt.tight_layout()
    plt.savefig(f"results/mpm/{names[i]}.png")
    plt.show()
```
